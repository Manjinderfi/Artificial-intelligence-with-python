# Arrays

#### Creating Arrays, size of Arrays

```python
# Examples for Tables, Vectors, Matrices, Graphs

import numpy as np

A = np.array([1,2,3,244,249091,1221])
print('A = ', A)
print(50*"=")

B = np.array([[1,2,44],[3,4,441]])
print('B = ', B)
print(50*"=")

print(np.shape(A))
print(np.shape(B))
```

#### zeros(), ones() and full()

```python
Z1 = np.zeros((15))
print("Z = ", Z1)
print(50*"=")

Z2 = np.zeros( (4,7)   )
print("Z2 = ",Z2)
print(50*"=")

I = np.ones((6,2))
print("I = ",I)
print(50*"=")

F = np.full((3,4),2.5)
print("F = ",F)
```

#### linspace() and arange()

```python
x = np.linspace(5,10,11)
print("x=",x)
print(50*"=")

y = np.arange(5,10,0.5)
print("y=",y)
print(50*"=")

# For comparison: range()
r = range(0,5,1)
for i in r:
    print(i)
```

#### Random

```python
# Random in numpy
dice = np.random.randint(1,7,30)
print("outcomes=", dice)
print(50*"=")

xn = np.random.randn(100)
print("xn=",xn)
print(50*"=")

x1 = np.random.random(25)
print("x1=",x1)
print(50*"=")
```

#### ndim and size, reshape, cutting and pasting

```python
# ndim ja size
B = np.array([[1,2,44],[3,4,441]])
print("B = ",B)
print("B.shape=",B.shape)
print("B.ndim=",B.ndim)
print("B.size=",B.size)
print(50*"=")

#  Reshape()
A1 = np.linspace(0,11,12)
print("A1 = ",A1)
print(A1.shape)
A2 = A1.reshape(2,6)
print("A2 = ",A2)
A3 = A2.reshape(3,4)
print("A3 = ",A3)
print(50*"=")

# Cutting/indexing Arrays
A = np.array([[1,2,3],[4,5,6]])
print("A=",A)
print('A[0,0] = ', A[0,0])
print('A[0,1] = ', A[0,1])
print('A[1,0] = ', A[1,0])
print('A[0,:] = ', A[0,:])
print('A[:,1] = ', A[:,1])
print(50*"=")

x = np.linspace(10,21,12)
print("x = ",x)
print('x[3:5] = ', x[3:5])
print('x[2:8:2] = ', x[2:8:2])
print(50*"=")

# Updating array by cutting and pasting
x[1] = 100
print("x = ",x)
A[1,1] = 99
print(" A = ",A)
A[1,:] = [97,98,99]
print("A = ",A)
A[1,:] = 101
print("A = ",A)
```

#### repeat and delete

```python
# Repeat()
row = [[1,2,3]]
A = np.repeat(row,4,axis=0)
print('A = ', A)
column = np.array([[1],[2],[3]])
print('column = ', column)

# Repeat
B = np.repeat(column,5,axis=1)
print('B = ', B)
print(50*"=")


A2 = np.vstack((A,A))
print("A2 = ",A2)
A3 = np.hstack((A,A))
print("A3 = ",A3)
print(50*"=")

# Delete()
print("A = ",A)
dlt1 = np.delete(A, [0],axis=0) # For one row, we can use 0 instead of [0]
print("After first delete = ",dlt)
dlt2 = np.delete(A,[0],axis=1)
print("After second delete = ",dlt2)
print(50*"=")
```

#### Iteration and reading text files into numpy

```python
# Iteration
A = np.array([1,2,3,4,5,6])
print("A before reshape: ",A)
print('')
A = A.reshape(2,3)
print("A after reshape: ",A)
print('')
n, m = np.shape(A)
print("n = ",n,"m = ",m)
for i in range(n):
    print("Row",i,"is",A[i,:])
for j in range(m):
    print("Column", j, "is", A[:, j])

for i in range(n):
    for j in range(m):
        print("Element", i,j, "is", A[i, j])

# Performing the same operation using nditer
for a in np.nditer(A):
    print("a = ",a)
print(50*"=")


#Reading file (CSV) (Comma Separated Values)
data = np.genfromtxt("chap2-data.csv",delimiter=",",skip_header=1)
print("data=",data)
print(50*"=")
```
